### Ejercicio 1.   Información de un empleado

Diseña un XML Schema que represente la información de un empleado. Cada empleado debería tener los siguientes detalles:

- EmpleadoID (numérico, requerido)
- Name (firstName y lastName, ambos requerido)
- Departamento (string, optional)
- Salary (numérico, opcional)

El nombre del schema debería ser TE54_1.xsd

El XML a validar es el siguiente:

```xml
<Employees xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance"
           xsi:noNamespaceSchemaLocation="TE54_1.xsd">

  <Employee>
    <EmployeeID>1</EmployeeID>
    <Name>
      <FirstName>John</FirstName>
      <LastName>Doe</LastName>
    </Name>
    <Department>IT</Department>
    <Salary>75000</Salary>
  </Employee>

  <Employee>
    <EmployeeID>2</EmployeeID>
    <Name>
      <FirstName>Jane</FirstName>
      <LastName>Smith</LastName>
    </Name>
    <Salary>60000</Salary>
  </Employee>

</Employees>
```

En el cuadro inferior incluye el XSD que lo valida.

### Ejercicio 2:  Catalogo de libros

Diseña un XML Schema que represente un catalogo de una librería.<br>
Cada libro debería tener los siguientes detalles:

- Un libro tiene un ISBN (atributo, string, requerido y único)
- Title (string, requerido)
- Author (string, requerido), tiene un atributo country (string, opcional y por defecto "EEUU")
- Genre (enumeración: Fiction, Non-Fiction,Mystery,Children, etc., opcional)
- PublicationYear (numérico, opcional)

El XML a validar es el siguiente:

```xml
<LibraryCatalog xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance"
                xsi:noNamespaceSchemaLocation="TE54_2.xsd">

  <Book ISBN="978-2-362-3432">
    <Title>The Great Gatsby</Title>
    <Author country="United kingdom">F. Scott Fitzgerald</Author>
    <ISBN>978-3-16-148410-0</ISBN>
    <Genre>Fiction</Genre>
    <PublicationYear>1925</PublicationYear>
  </Book>

  <Book ISBN="978-2-362-3432">
    <Title>Data Science Essentials</Title>
    <Author>Lisa Scientist</Author>
    <ISBN>978-1-23-456789-0</ISBN>
  </Book>

  <Book ISBN="978-1-23-456789-0">
    <Title>Amanda's First Day of School</Title>
    <Author country="France">Amanda Smith</Author>
    <Author>John Smith</Author>
    <Genre>Fiction</Genre>
    <Genre>Children</Genre>
    <PublicationYear>1997</PublicationYear>
  </Book>

  <Book ISBN="978-1-23-456789-0">
    <Title>Harry Potter and the Chamber of Secrets</Title>
    <Author country="United Kingdom">J. K. Rowling</Author>
    <Genre>Fiction</Genre>
    <PublicationYear>1998</PublicationYear>

</LibraryCatalog>
```

### Ejercicio 3. Itinerario de viaje

Diseña un XML Schema que represente un itinerario de un cruzero.<br>
Cada viaje debería tener los siguientes detalles:

- Un viaje tiene un código (atributo, string, requerido y único)
- Un viaje tiene un nombre (string, requerido)
- Un viaje tiene un destino (string, requerido)
- Un viaje tiene una fecha de salida (string, requerido)
- Un viaje tiene una fecha de llegada (string, opcional)
- Un viaje tiene un precio (numérico, opcional)
- Un viaje tiene una lista de actividades (obligratorio)
  - Cada actividad tiene un nombre (string, requerido)
  - Cada actividad tiene una descripción (string, opcional)
  - Cada acttividad tiene una fecha (date, obligatorio)
  - Cada actividad tiene un precio (numérico, opcional)
  - Cada actividad tiene un tipo (enumeración: "Aventura", "Cultural", "Gastronómica", "Relax", "Deportiva", "Otro", opcional)
  - Cada actividad tiene un número de plazas (entero, obligatorio)
  
- Cada viaje tiene una lista de excursiones (entero, obligatorio)
  - Tiene un nombre (string, requerido)
  - Tiene una descripción (string, opcional)
  - Tiene una fecha (date, obligatorio)
  - Tiene un precio (numérico, opcional)
  - Tiene una destino (string, obligatorio), el destino tiene un atributo tipo (enumeración: "Aventura", "Cultural", "Gastronómica", "Relax", "Deportiva", "Otro", opcional)
  - Tiene un número de plazas (entero, obligatorio)
  
- Cada viaje tiene una lista de pasajeros (entero, opcional)
  - Tiene un nombre (string, requerido)
  - Tiene un apellido (string, opcional)
  - Tiene una fecha de nacimiento (date, opcional)
  - Tiene un número de pasaporte (string, opcional)
  - Tiene un número de teléfono (opcional)
  - Tiene un email (string, opcional), con un atributo tipo (enumeración: "main", "secondary", opcional)
  - Tiene un número de habitación (string, opcional)
  - Tiene un tipo de habitación (enumeración: "Individual", "Doble", "Triple", opcional)
  - Número del billete (string, obligatorio)

El XML a validar es el siguiente:

Genera un XML que valide el siguiente XML Schema:

### Ejericio 7

Escribir un XML Schema para el siguiente documento XML, incluyen las restricciones necesarias para que se cumplan las siguientes condiciones:

- El elemento versión será una plantilla que permita números seguidos de un punto y seguido de más números.
- El elemento fecha será de tipo fecha.
- El elemento faq está compuesto por solo un elemento info y por uno o más elementos part.
- El elemento info está compuesto por un titulo, un autor, y puede tener o no email, versión y fecha.
- El elemento part puede tener uno o más elementos q.
- El elemento q está compuesto por los elementos qtext y a que son obligatorios.

```xml
<?xml version="1.0" encoding="UTF-8"?>
<faq xmlns:xsi="http://www.w3.org/2001/XMLSchema-instance" 
         xsi:noNamespaceSchemaLocation="TE54_7.xsd">
   <info>
      <titulo>WINDOWS98? SOLO SI ME OBLIGAN!</titulo>
      <autor>Fernando Damián Lorenzo García</autor>
      <email>Fer@yahoo.es</email>
      <version>1.0</version>
      <fecha>1999-05-01</fecha>
   </info>
   <part>
      <q>
         <qtext>¿Por qué tengo que instalar W95?</qtext>
         <a>A mi también me parece una buena pregunta.</a>
      </q>
      <q>
         <qtext>Y de que me vale?</qtext>
         <a>Otra ingeniosa pregunta.</a>
      </q>
   </part>
</faq>
```

En el cuadro inferior incluye el XSD que lo valida.
